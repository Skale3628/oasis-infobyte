```python
years=list(set(list(Unemployment_df[' Date'].dt.year)))
gbyyear_df=Unemployment_df.groupby(by='year')
mean_Unemployment_rate_overall =list(gbyyear_df[' Estimated Unemployment Rate (%)'].mean())
```

```python
sns.set_style("darkgrid")
plt.plot(years, mean_Unemployment_rate_overall, 's-b')
plt.xlabel('year')
plt.ylabel('mean_Unemployment_rate_overall(%)')
plt.title("Unemployment rate by Year")
```
